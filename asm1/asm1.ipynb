## Question 2: Missing values

The code below shows the number of employees with missing values for each column.

```python
print("Number of employees with missing values for each column:")
print(df2.isna().sum(axis=0))
print(len(df2))
```

For each variable with missing values, consider the following approaches to deal with them:
1.   Impute missing values with the mean.
2.   Impute missing values with the mode.
3.   Create a new value for observations with missing values.
4.   Drop observations with missing values.

Which of these do you think is the best way to proceed in this scenario? Explain your decision. **TIP:** Think carefully about the nature of the context, the variable(s) for which you are imputing values, and the purpose of your analysis.

Also, make sure to update the code below according to your chosen approach to deal with missing variables.

**Answer**: <font color='red'>
* I will choose to impute missing values with the new value like "UNKNOWN" which represent unknown satisfaction level.

* By replacing missing values with "Unknown," we acknowledge that the information about relationship satisfaction is missing for those specific employees. This approach preserves the fact that there is missing data without making assumptions or altering the original dataset.

* Avoiding bias: Replacing missing values with the mode or mean might introduce bias into the analysis, especially if the missing data is not missing completely at random. If the missingness is related to the employee's relationship satisfaction, using the mode or mean could distort the distribution and affect subsequent analyses.

*   Dropping the observation is not preferred as the missing rows = 177, account for 12 percent of the data, which is over 5%, thus being too large to drop.

*   Imputing with mean is also not recommended as the satisfaction is in categorical value. Given the data is right skewed, replacing missing data with mean would affect its distribution. its also not appropriate to replace categorical variable with mean (continuous value).  

 </font>

```python
import random

df3 = df2.copy()
# Replace the value of 'var' with the name of the variable with missing values.
var = "RelationshipSatisfaction"
# Replace the value of 'option' with the alternative you chose to deal with missing values.
option = 3

if option == 0:
    # This chooses a random option
    option = random.randint(1, 4)

if option == 1:
  # This is mean imputation
  df3[var] = df3[var].fillna(df3[var].mean())
elif option == 2:
  # This is mode imputation
  df3[var] = df3[var].fillna(df3[var].mode()[0])
elif option == 3:
  # This creates a new value for missing values
  # (you can replace the zero with something else if you want)
  new_val = "UNKNOWN"
  df3[var] = df3[var].fillna(new_val)
elif option == 4:
  # This drops observations with missing values
  df3 = df3[df3[var].isna() == False]

print(f"Option {option} was used to deal with missing values.")
```
